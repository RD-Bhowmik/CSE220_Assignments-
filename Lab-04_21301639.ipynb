```python
class Patient:
    def __init__(self, id, name, age, blood_group):
        self.id = id
        self.name = name
        self.age = age
        self.blood_group = blood_group
```

```python
class DoublyNode:
    def __init__(self, elem, next=None, prev=None):
        self.elem = elem
        self.next = next
        self.prev = prev
```

```python
class WRM:
    def __init__(self):
        self.head = DoublyNode(None)  # Dummy head
        self.head.next = self.head.prev = self.head


    def register_patient(self, id, name, age, blood_group):
        new_patient = Patient(id, name, age, blood_group)
        new_node = DoublyNode(new_patient)


        # about the tail
        new_node.next = self.head
        new_node.prev = self.head.prev
        self.head.prev.next = new_node
        self.head.prev = new_node


    def serve_patient(self):
        if self.head.next is self.head:  # No patients in the waiting room
            return None


        served_patient = self.head.next.elem
        self.head.next = self.head.next.next
        self.head.next.prev = self.head


        return served_patient


    def cancel_all(self):
        self.head.next = self.head.prev = self.head


    def can_doctor_go_home(self):
        return self.head.next is self.head


    def show_all_patients(self):
        current = self.head.next
        while current is not self.head:
            print(current.elem.id, end=" ")
            current = current.next
        print()




                                            #Driver code

waiting_room = WRM()


while True:
    print("\n1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All patients")
    print("4. Can Doctor go Home?")
    print("5. Cancel all Appointments")
    print("6. Exit")
    choice = int(input("Enter your choice: "))


    if choice == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")
        waiting_room.register_patient(id, name, age, blood_group)
        print("Patient added successfully.")
    elif choice == 2:
        served_patient = waiting_room.serve_patient()
        if served_patient:
            print("Patient being served:", served_patient.name)
        else:
            print("No patients in the waiting room.")
    elif choice == 3:
        print("All patients in the waiting room:")
        waiting_room.show_all_patients()
    elif choice == 4:
        if waiting_room.can_doctor_go_home():
            print("Yes, doctor can go home.")
        else:
            print("No, there are patients waiting.")
    elif choice == 5:
        waiting_room.cancel_all()
        print("All appointments canceled.")
    elif choice == 6:
        break
    else:
        print("Invalid choice. Please try again.")
```
